# ATM746 - Laboratory Component

## Step by step to get up to date on the lab

### Run the following commands sequentially without skipping steps

1. `cd $SCRATCH` -> go to your scratch folder

2. `mkdir -p atm746/ar_test` -> If the directory exists, an error is returned. Otherwise, it is created

3. `cd atm746/ar_test` -> go to this directory

4. `cp -r ~wrfhelp/derecho_pre_compiled_code/wpsv4.6.0 WPS` -> copy the WPS programs and files  (it takes some time)

5. `cd WPS/` -> go to this directory

6. `cp /glade/work/jhenao/atm746/WPS/namelist.wps /glade/derecho/scratch/$USER/atm746/ar_test/WPS` -> copy namelist

7. `cp /glade/work/jhenao/atm746/WPS/run*.sh /glade/derecho/scratch/$USER/atm746/ar_test/WPS` ` -> copy batch scripts to run programs

8. `qsub rungeogrid.sh` -> Submit the geogrid.exe job — this may take some time

9. `./link_grib.csh /glade/campaign/collections/rda/data/d083003/2024/202411/gdas1.fnl0p25.202411{1*,2*}.f00.grib2` -> link meteorological files

10. `ls -l GRIBFILE* | wc -l` -> should output `80` on the terminal; otherwise something went wrong

11. `ln -sf ungrib/Variable_Tables/Vtable.GFS Vtable` -> link FNL Vtable 

12. `qsub runungrib.sh` -> Submit the ungrib.exe job — this may take some time

13. `qstat -u $USER` -> Check the status of the submitted jobs. See below, geogrid is Runing (R) and ungrib is queued (Q) 

```
                                                                            Req'd  Req'd   Elap 
                Job ID          Username Queue    Jobname    SessID NDS TSK Memory Time  S Time 
                --------------- -------- -------- ---------- ------ --- --- ------ ----- - -----
                3268712.desche* jhenao   cpu      henao-geo* 189339   1   1  235gb 00:30 R 00:02
                3268787.desche* jhenao   cpu      henao-ung*    --    1   1  235gb 01:00 Q   -- 
```

14. After some time, confirm that both geogrid.exe and ungrib.exe ran successfully before proceeding with the next step

    - `ls -lh geo_em.d0* | wc -l` -> should output `2`

    - `ls -lh FILE* | wc -l`  -> should output `25`

    - `qstat -u $USER` -> you should see no output (no jobs running)

15. `qsub runmetgrid.sh ` -> Submit the metgrid.exe job — this may take some time

16. `qstat -u $USER` -> Check the status of the submitted jobs.

17. After some time, confirm that metgrid.exe ran successfully before proceeding with the next step

    - `ls -l met_em.d0* | wc -l` -> should output `50`

    - `qstat -u $USER` -> you should see no output (no jobs running)

18. `cd ..` -> go one directory up in hierarchy

19. `cp -r ~wrfhelp/derecho_pre_compiled_code/wrfv4.7.1  /glade/derecho/scratch/$USER/atm746/ar_test/WRF` -> copy the WRF programs and files (it takes some time)

21. `cd WRF/test/em_real/` -> go to directory 

23. `cp /glade/derecho/scratch/jhenao/atm746/ar_test/run*.sh .` -> copy batch scripts to run wrf programs

24. `cp /glade/derecho/scratch/jhenao/atm746/ar_test/namelist.input .` -> copy namelist

25. `ln -sf /glade/derecho/scratch/$USER/atm746/ar_test/WPS/met_em.d0* .` -> link met_em to create input files

26. `qsub runreal.sh` Submit the real.exe job to create input files — this may take some time - 

27. `qstat -u $USER` -> Check the status of the submitted jobs.

28. After some time, confirm that real.exe ran successfully before proceeding with the next step

    - `ls -l wrfinput_d0* | wc -l` -> should output `2`

    - `ls -l wrfbdy_d01 | wc -l` -> should output `1`

    - `qstat -u $USER` -> you should see no output (no jobs running)


29. Edit namelist.input to run your model configuration 

    - change `mp_physics` to your value (`Thompson -> 8`,`Morrison -> 10`, `WSM6 -> 6`)
    - change `bl_pbl_physics` to your value (`YSU -> 1`,`MYJ -> 2`, `BouLac -> 8`)
    - if your PBL option is `MYJ`, change `sf_sfclay_physics` to `2`

30. `qsub runwrf.sh` -> submit wrf job

31. `qstat -u $USER` -> Check the status of the submitted jobs.


32. when it is running, there are rsl.out and rsl.error files, one for each cpu used

    - you can check status there, for instance `tail rsl.error.0000` when finished should say `wrf: SUCCESS COMPLETE WRF`

    - you should see the hourly output files: `wrfout_d*...`
